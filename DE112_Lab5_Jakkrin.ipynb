# While statement

Objectives:

1. Learn how to use while and its relevant statements.

**1. While loop:** In while loop, the code inside is executed, if the condition is True and the condition is validated again after the code is executed. (The loop continues long as the condition is True.) The structure of while loop is:


```
while <condition>:
  statement1 # within while loop
  statement2 # within while loop
statement3 # outside while loop --executed after when while loop execution has ended

```




````python
a = 20
while (a>0):
  print(a)
  a = a-1
print('Loop has ended!')
````

**2. Nested while loop:** A nested loop is a loop inside a loop. The "inner loop" will be executed one time for each iteration of the "outer loop"

```
while <condition>: # outer loop
  statement1 # within outer loop
  statement2 # within outer loop
  while <condition>: # innner lop
    statement3 # within innner loop
    statement4 # within innner loop
  statement5 # within outer loop

statement6 # outside while loop --executed after when while loop execution has ended
```


````python
a = 5
while (a>0):
  print('a =',a)
  b=3
  while (b>0):
    print('b =',b)
    b =b-1
  a = a-1
print('Loop has ended!')
````

**While loop statement**
1. break: this statement is used to exit the current iterative loop  
2. continue:this statement is used to stop the current iteration and go back to condition checking statement. 
3. else: the code inside else block runs once the condition is false. This code block will not run when the break statement is used to exit the loop)


````python
# Illustration for break in while loop
a = 20
while (a>0):
  print(a)
  a = a-1
  if a == 10:
    break
print('Loop has ended!')
````

````python
# Illustration for continue in while loop
a = 20
while (a>0):
  a = a-1
  if a == 10:
    continue  
  print(a)
print('Loop has ended!')
````

````python
# Illustration for else in while loop
a = 9
while (a>0):
  a = a-1
  if a == 10:
    break  
  print(a)
else:
  print('a becomes zero or it does not start with positive number')

print('Loop has ended!')
````

````python
# Illustration for if-statement inside while loop
a = 20
while (a>0):
  a = a-1
  if a == 10:
    pass  
  print(a)
else:
  print('a becomes zero or it does not start with positive number')

print('Loop has ended!')
````

**Infinite loop**: Infinite loop is the loop that runs forever. 
- While loop can be made an infinite loop with the condition that always true 
- while break statement can be used to exit the infinite loop
```
while (True):
    user_input = input('Your favorite number is:')
    print(user_input)
    if (user_input == 0):
      break
```



````python
# Illustration for infinite loop
while (True):
  user_input = input('Your favorite number is:')
  print(user_input, type(user_input))
  if (int(user_input) == 0):
      break
print('Loop has ended!')
````
